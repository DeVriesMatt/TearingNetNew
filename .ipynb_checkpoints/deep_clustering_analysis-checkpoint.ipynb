```python
from sklearn import manifold
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import umap

def calc_pred(mod, loader):
    outputs_t = []
#     features_t = []
    embeddings_t = []
    clusterings_t = []
    labels_t = []
    mod.eval()
    for data in tqdm(loader):
        with torch.no_grad():
            pts, lab, _, _ = data
#             labels_t.append(lab.detach().numpy())
            inputs = pts.to(device)
            output, feature, clustering_out = mod(inputs)
            outputs_t.append(torch.squeeze(output).cpu().detach().numpy()) 
#             features_t.append(torch.squeeze(feature).cpu().detach().numpy()) 
            embeddings_t.append(torch.squeeze(feature).cpu().detach().numpy()) 
            clusterings_t.append(torch.squeeze(clustering_out).cpu().detach().numpy()) 
               
    output_array = np.asarray(clusterings_t)
    preds = np.argmax(output_array.data, axis=1)
    features = np.asarray(embeddings_t)
#     scalar = StandardScaler()
#     scaled_features = scalar.fit_transform(features)
    print("Calculating embedding")
    reducer = umap.UMAP()
# embedding = reducer.fit_transform(scaled_features)
    embedding = reducer.fit_transform(features)
    
    print('Embedding calculation done!')
    b = np.zeros((len(embedding), 3))
    b[:, 0] = embedding[:, 0]
    b[:, 1] = embedding[:, 1]
    b[:, 2] = kmeans.labels_
    data = pd.DataFrame(b, columns=['Umap1','Umap2', 'label'])
    facet_tsne = sns.lmplot(data=data, x='Umap1', y='Umap2', hue='label',
                       fit_reg=False,
                       legend=True,
                       scatter_kws={"s": 6})
    plt.show()
    plt.close()
    
    return output_array, 1, preds, features, labels_t
```

```python
from encoders.dgcnn import ChamferLoss

criterion_rec = ChamferLoss()
criterion_clust = torch.nn.KLDivLoss(reduction='sum')
```

```python
# Calculate target distribution
def target(out_distr):
    tar_dist = out_distr ** 2 / np.sum(out_distr, axis=0)
    tar_dist = np.transpose(np.transpose(tar_dist) / np.sum(tar_dist, axis=1))
    return tar_dist
```

```python
import numpy as np


def check_tolerance(cluster_predictions, previous_cluster_predictions):
    delta_label = (
        np.sum(cluster_predictions != previous_cluster_predictions).astype(
            np.float32
        )
        / cluster_predictions.shape[0]
    )
    previous_cluster_predictions = np.copy(cluster_predictions)
    return delta_label, previous_cluster_predictions
```

```python
optimizer = torch.optim.Adam(model.parameters(), 
                             lr=0.000001*16/batch_size, 
                             betas=(0.9, 0.999), 
                             weight_decay=1e-6)
```

```python
dataloader = DataLoader(dataset, batch_size=16, shuffle=False)
dataloader_inf = DataLoader(dataset, batch_size=1, shuffle=False)
model.to(device)
```

```python
from torch.autograd import Variable
from torch.utils.tensorboard import SummaryWriter
import os
import datetime
from tqdm import tqdm

writer = SummaryWriter()
date_time = str(datetime.datetime.now()).replace(" ", "_").replace("-", "_")[:-7]
num_epochs = 200
total_loss = 0.
rec_loss = 0.
clus_loss = 0.
model.train()
threshold = 0.
losses = []
test_acc = []
best_acc = 0.
divergence_tolerance = 1e-3

# initialise target distribution
print("Initialising target distribution")
# output_distribution, _, preds_prev, _, l = calc_pred(
#         model, dataloader_inf
#     )

# tar_distribution = target(output_distribution)

for epoch in range(num_epochs):
    print('Training epoch {}'.format(epoch))
    batch_num = 1
    if (epoch % 5 == 0) and (epoch != 0):
      
        print('Caluculating target distribution')
        output_distribution, _, preds, _, l = calc_pred(model, dataloader_inf)
        print(output_distribution.shape)
        
        tar_distribution = target(output_distribution)
        
        print('Checking Tolerance')
        delta_label, preds_prev = check_tolerance(
                preds, preds_prev
            )
        if delta_label < divergence_tolerance:
            print(
                    f"Label divergence {delta_label} < "
                    f"divergence tolerance {divergence_tolerance}"
                )
            print("Reached tolerance threshold. Stopping training.")
            break
        
        
    print('Training epoch {}'.format(epoch))
    model.train()
    batches = []
    for i, data in enumerate(dataloader, 0):
        inputs, labels, _, _ = data
        inputs = inputs.to(device)
        
        # ===================forward=====================
        with torch.set_grad_enabled(True):
            output, feature, clustering_out = model(inputs)
            optimizer.zero_grad()
            loss_rec = criterion_rec(inputs, output)
            tar_dist = torch.from_numpy(tar_distribution[((batch_num - 1) * batch_size):(batch_num*batch_size), :]).unsqueeze(1).to('cuda:0')
            loss_clus = criterion_clust(torch.log(clustering_out.unsqueeze(1)), tar_dist)

            loss = loss_rec + 1000 * loss_clus
#             loss = loss_clus
            # ===================backward====================
            loss.backward()
            optimizer.step()
            
        batch_num += 1
#         losses.append(float(loss.cpu()))
        writer.add_scalar('/Loss' + 'Batch', loss.item()/batch_size, (i + 1)*(epoch+1))
        lr = np.asarray(optimizer.param_groups[0]['lr'])

        if i % 10 == 0:
            print('[%d/%d][%d/%d]\tLossTot: %.4f\tLossRec: %.4f\tLossClus: %.8f'
                  % (epoch, num_epochs, i, len(dataloader), loss.item()/batch_size,
                     loss_rec.item()/batch_size, loss_clus.item()/batch_size,))

    # ===================log========================
    total_loss += float(loss)
    
    print('epoch [{}/{}], loss:{}'
          .format(epoch+1, num_epochs, total_loss))
    
    total_loss += float(loss)
    rec_loss += float(loss_rec)
    clus_loss += float(loss_clus)
    print('epoch [{}/{}], loss:{:.4f}, Rec loss:{:.4f}, Clust loss:{:.4f}'
          .format(epoch+1, num_epochs, total_loss, rec_loss, clus_loss))

    
torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': total_loss
            }, '128cell_3clus_cell.pt')
```
