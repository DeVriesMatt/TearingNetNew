```python
ae = GraphAutoEncoder(num_features=128, k=20, encoder_type="dgcnn", decoder_type='foldingnet')
model = DEC(autoencoder=ae, num_clusters=3)
checkpoint = torch.load(path)

model.load_state_dict(checkpoint['model_state_dict'])
```

```python
root_dir ='/home/mvries/Documents/Datasets/OPM/SingleCellFromNathan_17122021/'
df = '/home/mvries/Documents/Datasets/OPM/SingleCellFromNathan_17122021/all_cell_data.csv'
dataset = PointCloudDatasetAllDistalRotation(df, root_dir,
                              img_size=400,
        label_col="Treatment",
        transform=None,
        target_transform=None,
        cell_component="cell")

dataloader = DataLoader(dataset, batch_size=1, shuffle=False)
```

```python
from tqdm import tqdm
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()
all_feat = []
all_outputs = []
all_inputs = []
all_rotated_inputs = []
all_clusters = []

all_rot_feat = []
all_rot_outputs = []
all_rotated_inputs = []
all_rot_clusters = []

labels = []
all_serial_numbers = []
for data in tqdm(dataloader):
    inputs = data[0]
    lab = data[1]
    rotated_input = data[2].to(device)
    serial_num = data[3]
    inputs = inputs.to(device)
    batch_size = inputs.shape[0]

    output, features, clusters = model(inputs)
    rotated_output, rotated_features, rotated_clusters = model(rotated_input)
    all_inputs.append(torch.squeeze(inputs).detach().cpu().numpy())
    all_outputs.append(torch.squeeze(output).detach().cpu().numpy())
    all_feat.append(torch.squeeze(features).detach().cpu().numpy())
    all_clusters.append(torch.squeeze(clusters).detach().cpu().numpy())
    all_rotated_inputs.append(torch.squeeze(rotated_input).detach().cpu().numpy())
    
    all_rot_outputs.append(torch.squeeze(rotated_output).detach().cpu().numpy())
    all_rot_feat.append(torch.squeeze(rotated_features).detach().cpu().numpy())
    all_rot_clusters.append(torch.squeeze(rotated_clusters).detach().cpu().numpy())
    
    labels.append(lab[0])
    all_serial_numbers.append(serial_num[0])
```

```python
import umap
from sklearn import manifold
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scaled_features = scalar.fit_transform(all_feat)
# UMAP for vizualization
reducer = umap.UMAP(random_state=42)
# embedding = reducer.fit_transform(scaled_features)
embedding = reducer.fit_transform(scaled_features)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline

b = np.zeros((len(embedding), 2))
b[:, 0] = embedding[:, 0]
b[:, 1] = embedding[:, 1]

data = pd.DataFrame(b, columns=['Umap1','Umap2'])
data['label'] = np.asarray(labels)


facet = sns.lmplot(data=data, x='Umap1', y='Umap2', hue='label',
                   fit_reg=False, legend=True, legend_out=True, scatter_kws={"s": 6})


plt.show()
```

```python
xy = pd.DataFrame(scaled_features)
xy['label'] = data['label']
blebb_noc = xy[(xy['label'] == 'Nocodazole')|
              (xy['label'] == 'Blebbistatin')]
```

```python
blebb_noc
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import plot_confusion_matrix


X_train, X_test, y_train, y_test = train_test_split(blebb_noc.iloc[:,:-1], 
                                                        blebb_noc['label'],
                                                        random_state=0, 
                                                        test_size=0.33)
    
clf = svm.SVC(kernel='linear', C=1, random_state=42, class_weight='balanced')
clf.fit(X_train, y_train)


print(clf.score(X_train, y_train))
```

```python
print(clf.score(X_test, y_test))
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
cm = plot_confusion_matrix(clf, X_test, y_test, normalize='true', ax=ax, cmap=plt.cm.plasma)  
cm.ax_.get_images()[0].set_clim(0, 1)
plt.show()
```
